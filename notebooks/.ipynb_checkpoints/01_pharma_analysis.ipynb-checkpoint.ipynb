```python
df.columns = [
    'medicine_name',
    'composition',
    'uses',
    'side_effects',
    'image_url',
    'manufacturer',
    'excellent_review_pct',
    'average_review_pct',
    'poor_review_pct'
]
```

```python
df.columns
```

```python
import numpy as np

np.random.seed(42)

df['selling_price'] = np.round(
    np.random.uniform(50, 500, len(df)), 2
)
```

```python
df['purchase_price'] = np.round(
    df['selling_price'] * np.random.uniform(0.6, 0.8, len(df)), 2
)

df['profit'] = np.round(
    df['selling_price'] - df['purchase_price'], 2
)

df['profit_margin_pct'] = np.round(
    (df['profit'] / df['selling_price']) * 100, 2
)
```

```python
df['trust_score'] = (
    df['excellent_review_pct'] * 1.0 +
    df['average_review_pct'] * 0.5 -
    df['poor_review_pct'] * 1.0
)
```

```python
df['monthly_sales_qty'] = np.random.poisson(lam=30, size=len(df))
```

```python
from datetime import datetime, timedelta

df['expiry_date'] = [
    datetime.today() + timedelta(days=int(x))
    for x in np.random.randint(30, 720, len(df))
]
```

```python
df['days_to_expiry'] = (df['expiry_date'] - datetime.today()).dt.days

df['expiry_risk'] = pd.cut(
    df['days_to_expiry'],
    bins=[0, 30, 90, 10000],
    labels=['High Risk', 'Medium Risk', 'Low Risk']
)
```

```python
df['stock_movement'] = pd.cut(
    df['monthly_sales_qty'],
    bins=[0, 10, 40, 1000],
    labels=['Slow Moving', 'Medium Moving', 'Fast Moving']
)
```

```python
df['competitor_price'] = np.round(
    df['selling_price'] * np.random.uniform(0.9, 1.15, len(df)), 2
)
```

```python
df['price_position'] = np.where(
    df['selling_price'] > df['competitor_price'],
    'Overpriced',
    'Competitive'
)
```

```python
import os

os.makedirs("../outputs", exist_ok=True)

```

```python
df.to_csv("../outputs/retail_pharmacy_analytics.csv", index=False)
```
